```python
# prompt: Terapkan threshold untuk mengekstrak area hitam

from google.colab import drive
import numpy as np
from PIL import Image

drive.mount('/content/drive')

def process_png(file_path, threshold=128):
  """
  Loads a PNG image, applies a threshold to extract black areas, and saves the result.

  Args:
      file_path: The path to the PNG file.
      threshold: The threshold value (0-255) for black area extraction.
  """
  try:
    img = Image.open(file_path).convert("L")  # Directly convert to grayscale
    img_array = np.array(img)

    # Apply threshold: Pixels below the threshold become black (0), others become white (255)
    black_areas = (img_array < threshold) * 255
    black_areas_img = Image.fromarray(black_areas.astype(np.uint8))

    # Save the image with black areas extracted
    output_path = '/content/drive/MyDrive/black_areas_image.png'
    black_areas_img.save(output_path)

    print("Image shape:", img_array.shape)
    print(f"Black areas extracted using threshold {threshold}.")
    print(f"Image with black areas saved to {output_path}")

  except FileNotFoundError:
    print("File not found:", file_path)
  except Exception as e:
    print("An error occurred:", e)


# Example usage:
file_path = '/content/drive/MyDrive/kuliah math industri/_BB_kategori_IMG17225709061576 - Copy - Copy (1).png'
# Adjust the threshold as needed (e.g., 50, 100, 150)
process_png(file_path, threshold=50)
```
