```python
# prompt: tuliskan  deret Fibonacci

def fibonacci(n):
  """
  This function calculates the Fibonacci sequence up to n terms.
  """
  if n <= 0:
    return []
  elif n == 1:
    return [0]
  else:
    list_fib = [0, 1]
    while len(list_fib) < n:
      next_fib = list_fib[-1] + list_fib[-2]
      list_fib.append(next_fib)
    return list_fib

# Example usage:
n_terms = 10  # Change this to the desired number of terms
fib_sequence = fibonacci(n_terms)
fib_sequence
```

```python
# prompt: tuliskan rumus volume balok

# Rumus volume balok: panjang x lebar x tinggi

def volume_balok(panjang, lebar, tinggi):
  """Menghitung volume balok.

  Args:
    panjang: Panjang balok.
    lebar: Lebar balok.
    tinggi: Tinggi balok.

  Returns:
    Volume balok.
  """
  return panjang * lebar * tinggi

# Contoh penggunaan
panjang = 5
lebar = 3
tinggi = 2
volume = volume_balok(panjang, lebar, tinggi)
print(f"Volume balok dengan panjang {panjang}, lebar {lebar}, dan tinggi {tinggi} adalah: {volume}")
```

```python
# prompt: tuliskan Konversi Sudut (Radian dan Derajat)

import math

def derajat_ke_radian(derajat):
  """Mengkonversi sudut dari derajat ke radian.

  Args:
    derajat: Sudut dalam derajat.

  Returns:
    Sudut dalam radian.
  """
  return math.radians(derajat)

def radian_ke_derajat(radian):
  """Mengkonversi sudut dari radian ke derajat.

  Args:
    radian: Sudut dalam radian.

  Returns:
    Sudut dalam derajat.
  """
  return math.degrees(radian)
# Contoh penggunaan
sudut_derajat = 45
sudut_radian = derajat_ke_radian(sudut_derajat)
print(f"{sudut_derajat} derajat sama dengan {sudut_radian} radian")

sudut_radian = math.pi / 2
sudut_derajat = radian_ke_derajat(sudut_radian)
print(f"{sudut_radian} radian sama dengan {sudut_derajat} derajat")
```

```python
# prompt: gambarkan sin x dari x=0 sampai x=2 pi

import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 2*pi
x = np.linspace(0, 2 * np.pi, 100)

# Calculate sin(x)
y = np.sin(x)

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Plot of sin(x) from 0 to 2π")

# Add grid
plt.grid(True)

# Display the plot
plt.show()
```
